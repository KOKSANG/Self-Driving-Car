## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=5):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    # line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    # draw_lines(line_img, lines)
    return lines

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.9, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

## Creating color mask for images

The image is first converted into "HLS" as it gives the best edges detection results which will be justified after this.

Lane lines mainly consist of two colours, **YELLOW** and **WHITE** and we are going to only detect these colors on the road. The lower and upper threshold for both white and yellow can be referenced from a HLS color chart.

I use **cv2.inRange** to filter white and yellow individually before combining the masks by stacking them using **cv2.bitwise_or**.

````python
def convert_image(image):
    return cv2.cvtColor(image, cv2.COLOR_RGB2HLS)

def get_color_mask(image):
    converted_image = convert_image(image)
    
    # white mask
    lower = np.uint8([  0, 210,   0])
    upper = np.uint8([255, 255, 255])
    white_mask = cv2.inRange(converted_image, lower, upper)
    
    # yellow color mask
    lower = np.uint8([ 20,   0, 110])
    upper = np.uint8([ 45, 255, 255])
    yellow_mask = cv2.inRange(converted_image, lower, upper)
    
    # combine the mask for yellow and white
    combined_mask = cv2.bitwise_or(white_mask, yellow_mask)
    
    return cv2.bitwise_and(image, image, mask = combined_mask)
````

## Define region of interest

To capture only what we need from the road, I make a polygon shaped region that only the edges detected inside this region are taken for further processing.

````python
def get_region_of_interest(image, edges):
    ysize = image.shape[0]
    xsize = image.shape[1]
    
    # defining left, right and apex points for roi vertices
    left_point = [xsize/10, ysize]
    right_point = [xsize/10*9, ysize]
    apex1 = [xsize/9*4, ysize*0.6]
    apex2 = [xsize/9*5, ysize*0.6]
    
    # make vertices
    vertices = np.array([[left_point, apex1, apex2, right_point]], dtype=np.int32)
    
    # only getting edges inside the vertices
    roi_edges = region_of_interest(edges, vertices)
    
    return roi_edges
````

## Get hough lines transform

Then, we can use hough lines transform to detect lines in the edges. In this case, Probabilistic Hough Transform, **cv2.HoughLinesP** is used with a few optimum parameters pre-defined:
1. rho (p): distance resolution in pixels of the Hough grid
2. theta (): angular resolution in radians of the Hough grid
3. threshold: minimum number of votes or intersection points in Hough grid cell
4. min_line_length: minimum number of pixels making up a line, shorter will be skipped
5. max_line_gap: maximum gap in pixels between connectable line segments, pixels with bigger gap are ignored

At the end, only the lines in our ROI edges are retained.

````python
def get_roi_hough_lines(roi_edges):
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/360 # angular resolution in radians of the Hough grid
    threshold = 30   # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 40 # minimum number of pixels making up a line
    max_line_gap = 80 # maximum gap in pixels between connectable line segments
    lines_mask = hough_lines(roi_edges, rho, theta, threshold, min_line_length, max_line_gap)
    
    return lines_mask
````

## Interquartile range

Here, iqr range is used to eliminate outliers that happen to be skewed away from median. In our case here, only the median of our x coordinates are considered as y coordinates for the lines we draw are always consistent.

Our "lines" argument for **get_iqr_range** consists of an array of list [x1, slope, y-intercept, line length].

Notes:
1. Percentile range is 25 to 75
2. Only x1, our first x coordinate in each line is tested for skewness

````python
# interquartile range
def get_iqr_range(lines, rng=(25, 75)):
    # only getting iqr range of their first x, x1
    _iqr = iqr(lines[:, 0], rng=rng)
    
    # 1st and 3rd quartile
    q1, q3 = np.percentile(lines[:, 0], rng)
    
    # lower and upper limit
    lower = q1 - (_iqr*1.5)
    upper = q3 + (_iqr*1.5)
    
    # only return lines which are inside lower and upper limit
    return (lines[np.where((lines[:, 0] >= lower) & (lines[:, 0] <= upper))])
````

## Get left and right lines on lane

To get left and right lines, all the lines collected have to be separated into either left or right lines. **np.polyfit** is used to get the slope, m and y-intercept, c for each line.

Then, lines with skewed x1 are removed using **get_iqr_range**. Next, the average of all respective lines are calculated, results are in the form of average slope and average y-intercept.

To calculate their average, **np.dot** is used to form dot products between the line length and their m and c. Then, they are all divided by their summation of length. This is to smoothen out the skewed lines.

````python
# get the average left and right lines
def get_left_right_lines(image, lines):
    ysize = image.shape[0]
    xsize = image.shape[1]
    left_lines = []
    right_lines = []
    
    # goes through each line in our collection of lines
    for line in lines:
        for x1, y1, x2, y2 in line:
            m, c = np.polyfit((x1, x2), (y1, y2), 1) # calculate their slope and y-intercept
            length = np.absolute(np.sqrt((x2-x1)**2 + (y2-y1)**2)) # calculate their line length
            
            if x1 < xsize/2: # if x1 < half of xsize, it is left
                left_lines.append([x1, m, c, length])
                
            elif x1 > xsize/2: # if x1 > half of xsize, it is right
                right_lines.append([x1, m, c, length])
    
    # make left and right lines collection into numpy array
    left_lines = np.array(left_lines)
    right_lines = np.array(right_lines)
    
    # filter left and right lines that are out of the interquartile range
    left_lines = get_iqr_range(left_lines)
    right_lines = get_iqr_range(right_lines)
    
    all_left_lines = left_lines # all_left_lines is made of list of line's (x1, slope, y-intercept, length)
    all_right_lines = right_lines # same with all_left_lines but for right
    
    # dot product of length and (m, c)
    # also divide them by summation of lines length
    # output are average m and c of all left and right lines
    path_left = np.dot(all_left_lines[:, -1], all_left_lines[:, 1:-1])/ np.sum(all_left_lines[:, -1]) 
    path_right = np.dot(all_right_lines[:, -1], all_right_lines[:, 1:-1])/ np.sum(all_right_lines[:, -1])
    
    return path_left, path_right  
````

## Draw or extrapolate lines

Get the average lines as input, then determines whether they are at a left, right turn or straight line by comparing their left and right slopes.

Define the max and min range of y coordinate that we want to draw on the image. Calculate their x-intercept afterwards using the m and c calculated from the lines.

Draw the lines!

Notes:
1. Left top corner of image is our (0, 0)
2. x axis increases from left to right, y axis increases from top to bottom
3. For left turn, left and right slope have to be positive, negative for right turn.
4. And for straight line, left slope is negative and right slope otherwise. 

````python
# to draw lines on the image
def draw_left_right_lines(image, lines):
    output_image = np.copy(image)
    ysize = image.shape[0]
    xsize = image.shape[1]
    
    # getting max and min of drawing y coordinate
    ymax = int(ysize)
    ymin = int(ysize*0.63)
    
    # getting the average left and right lines
    # each of them consists of their slope and intercept
    path_left, path_right = get_left_right_lines(image, lines)
    
    # get the slope and intercept
    m_left, c_left = path_left
    m_right, c_right = path_right
    
    # left turn
    if m_left > 0 and m_right > 0:
        left_xy1 = (int((ymin - c_left)/m_left), ymin) # if it is left turn, left x1 is our min coordinates
        left_xy2 = (int((ymax - c_left)/m_left), ymax) # left x2 is max coordinates
        right_xy1 = (int((ymin - c_right)/m_right), ymin) # right x1 is our min coordinates
        right_xy2 = (int((ymax - c_right)/m_right), ymax) # right x2 is max coordinates
    
    # right turn
    elif m_left < 0 and m_right < 0:
        left_xy1 = (int((ymax - c_left)/m_left), ymax) # if right turn, left x1 is max
        left_xy2 = (int((ymin - c_left)/m_left), ymin) # left x2 is min
        right_xy1 = (int((ymax - c_right)/m_right), ymax) # it goes the same for right x1, x2
        right_xy2 = (int((ymin - c_right)/m_right), ymin)
        
    # straight line
    elif m_left < 0 and m_right > 0:
        left_xy1 = (int((ymax - c_left)/m_left), ymax) # if straight, left x1 is max
        left_xy2 = (int((ymin - c_left)/m_left), ymin) # left x2 is min
        right_xy1 = (int((ymin - c_right)/m_right), ymin) # right x1 is min
        right_xy2 = (int((ymax - c_right)/m_right), ymax) # right x2 is max
    
    # draw left and right lines based of previous coordinates calculated
    cv2.line(output_image, left_xy1, left_xy2, (255, 0, 0), 10)
    cv2.line(output_image, right_xy1, right_xy2, (255, 0, 0), 10)
    
    return output_image 
````

This is to test for all images. All images are fed into a pipeline of lane lines finding.

The pipeline goes as below:
1. Read image
2. Convert image into HLS scale
3. Convert HLS image into gray scale
4. Gaussian blur the image to reduce noise
5. Find canny edges
6. Filter edges outside of the ROI
7. Hough transform the edges in our ROI
8. Collect all lines, average all the lines and draw them

````python
i = 0
plt.figure(figsize=(15, 25))

for item in images:
    image = mpimg.imread('test_images/' + item)
    color_mask= get_color_mask(image)
    gray_image = grayscale(color_mask)
    gaussian_image = gaussian_blur(gray_image, kernel_size=11)
    edges = canny(gray_image, low_threshold = 100, high_threshold = 200)
    roi_edges = get_region_of_interest(image, edges)
    lines = get_roi_hough_lines(roi_edges)
    output_image = draw_left_right_lines(image, lines)
    
    i += 1

    plt.subplot(4, 2, i)
    plt.imshow(output_image)
    
plt.tight_layout(pad=0, h_pad=0, w_pad=0)
plt.show()
````
