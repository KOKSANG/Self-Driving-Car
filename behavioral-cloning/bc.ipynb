```python
import os
import glob
import zipfile
import pathlib
import cv2
import skimage as sk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from sklearn.model_selection import  train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.utils import shuffle
```

```python
# create grayscale image
def grayscale(img):
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# normalize image to zero mean
def normalize(img):
    mean = np.mean(img)
    std = np.std(img)
    return (img-mean)/std
# augment image, left right flip for now
def augment(image, randn):
    return np.flip(image, axis=randn%2).astype(np.uint8)
```

```python
df_c = pd.concat([df.center, df.steering], axis=1).rename(index=str, columns={'center': 'img'})
df_l = pd.concat([df.left, df.steering], axis=1).rename(index=str, columns={'left': 'img'})
df_r = pd.concat([df.right, df.steering], axis=1).rename(index=str, columns={'right': 'img'})
```

```python
def yeo_johnson_bias(steering):
    if steering >= 0:
        return np.log(steering + 1)
    elif steering < 0:
        return -np.log(-steering + 1)
```

```python
for index, value in df_l.iterrows():
    df_l.loc[index, 'steering'] = df_l.loc[index, 'steering'] + yeo_johnson_bias(df_l.loc[index, 'steering'])
```

```python
for index, value in df_r.iterrows():
    df_r.loc[index, 'steering'] = df_r.loc[index, 'steering'] + yeo_johnson_bias(df_r.loc[index, 'steering'])
```

```python
df_r[df_r.steering < 0].head()
```

```python
new_df = pd.concat([df_c, df_l, df_r], axis=0)
```
